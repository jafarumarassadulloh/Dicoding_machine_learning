## Modeling

menentukan input-output

```python
features = ['Engine Size(L)','Cylinders','Fuel Consumption Comb (L/100 km)']
```

```python
target = ['CO2 Emissions(g/km)']
```

```python
X = data1[features]
y = data1[target]
```

Split data Train dan Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print(f'Total # of sample in whole dataset: {len(X)}')
print(f'Total # of sample in train dataset: {len(X_train)}')
print(f'Total # of sample in test dataset: {len(X_test)}')
```

membuat model Regresi Linier

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_prediction = regressor.predict(X_test)
print(y_prediction[:5])
print('*'*40)
print(y_test[:5])
```

```python
y_test.describe()
```

```python
RMSE = sqrt(mean_squared_error(y_true=y_test, y_pred=y_prediction))
print('Training Accuracy: ', regressor.score(X_train, y_train)*100)
print('Testing Accuracy: ', regressor.score(X_test, y_test)*100)
print('Root Mean Squared Error: ', RMSE)
```

membuat model Random Forest



```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
import math
rf = RandomForestRegressor(random_state = 42)
rf.fit(X_train, y_train)
pred = rf.predict(X_test)
mse = mean_squared_error(y_test, pred)
rmse = math.sqrt(mse)
print('Training Accuracy: ', rf.score(X_train, y_train)*100)
print('Testing Accuracy: ', rf.score(X_test, y_test)*100)
print('Root Mean Squared Error: ', rmse)
```

membandingkan kedua model dengan plot



```python
# Data dari dua model
scores_model1 = [regressor.score(X_train, y_train) * 100, regressor.score(X_test, y_test) * 100, RMSE]
scores_model2 = [rf.score(X_train, y_train) * 100, rf.score(X_test, y_test) * 100, rmse]
labels = ['Training Score', 'Test Score', 'RMSE']

# Membuat bar plot
bar_width = 0.35
index = np.arange(len(labels))

plt.bar(index, scores_model1, bar_width, label='Regresi Linier', color='blue', alpha=0.7)
plt.bar(index + bar_width, scores_model2, bar_width, label='Random Forest', color='red', alpha=0.7)

# Menambahkan label sumbu dan legenda
plt.xlabel('Model')
plt.ylabel('Nilai')
plt.xticks(index + bar_width / 2, labels)
plt.legend(loc='upper right')

# Menambahkan judul
plt.title('Perbandingan Performa antara Regresi Linier dengan Random Forest')

# Menampilkan bar plot
plt.tight_layout()
plt.show()
```

membandingkan kedua model dengan tabel

```python
# Membuat DataFrame
df = pd.DataFrame({'Regresi Linier': scores_model1,
                   'Random Forest': scores_model2,
                   'Evaluasi': ['Akurasi Training Score','Akurasi Tes Score', 'RMSE']})

# Menampilkan DataFrame sebagai tabel
print(df)
```
